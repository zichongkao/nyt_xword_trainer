```python
import pandas as pd
```

```python
df = pd.read_csv('./NYT Crossword_2009_2016.csv',  na_filter=False) # There are some clues with "NULL" as the answer.
```

```python
df.head()
```

```python
count_df = pd.DataFrame(df.groupby('Word')['Clue'].count())
count_df.columns = ['Answer Count']
count_df['Answer Rank'] = count_df['Answer Count'].rank(method='min', ascending=False)
```

```python
count_df.sort_values('Answer Count', ascending=False)
```

```python
big_df = pd.merge(
    df,
    count_df,
    how="left",
    on="Word",
    copy=True,
    indicator=False,
    validate=None,
)
big_df.head()
```

```python
big_df.tail()
```

```python
display(big_df.dtypes)
big_df[big_df['Answer Count'].isna()]
```

```python
big_df['Answer Count'] = big_df['Answer Count'].astype(int)
big_df['Answer Rank'] = big_df['Answer Rank'].astype(int)
big_df = big_df[['Year', 'Weekday', 'Clue', 'Word', 'Explanation', 'Answer Count', 'Answer Rank']]
```

```python
big_df.head()
```

```python
big_df.to_csv('./NYT Crossword_2009_2016_processed.csv', index=False)
```
